````python
# Write your code to make your dataset analysis ready.
#First step will be cleaning the data as it has lots of dupicate values and missing/duplicate values.
Play_Store_data=Play_Store_data.drop_duplicates('App')
Play_Store_data
````

````python
#First step will be cleaning the data as it has lots of dupicate values and missing/duplicate values.
reviews=reviews.drop_duplicates('Translated_Review')
reviews
````

````python
#Now we will find null value if present and then replace it.
Play_Store_data.isnull().sum()
````

````python
#using mean imputation
playstore_mean=Play_Store_data['Rating'].mean()
playstore_mean = round(playstore_mean, 1)
playstore_mean
Play_Store_data['Rating']=Play_Store_data['Rating'].fillna(playstore_mean)
Play_Store_data
````

````python
Play_Store_data.isnull().sum()
````

````python
Play_Store_data['Type']=Play_Store_data['Type'].fillna('Unknown')
````

````python
Play_Store_data['Android Ver']=Play_Store_data['Android Ver'].fillna('Unknown version')
````

````python
Play_Store_data['Current Ver']=Play_Store_data['Current Ver'].fillna('Unknown version')
````

````python
Play_Store_data['Content Rating']=Play_Store_data['Content Rating'].fillna('Unknown')
````

````python
Play_Store_data.isnull().sum()
````

````python
#Converting datatype of Prices
Play_Store_data['Price'] = Play_Store_data['Price'].str.replace('$', '').astype(float)
````

````python
Play_Store_data['Installs'] = Play_Store_data['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
````
